```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
rundir = 'out/17/'
```

```python
err_df = pd.read_csv(f'{rundir}nrmse.csv')
plt.plot(err_df['err_train'], label='nrmse_train')
plt.plot(err_df['err_val'], label='nrmse_val')
plt.yscale('log')
plt.legend()
plt.show()
```

```python
err_best = err_df['err_val'].min()
print(f'Best NRMSE: {err_best*100:.3}%')
```

```python
lp = pd.read_csv(f'{rundir}predictions.csv')
plt.plot([0,10000], [0,10000], color='gray')
plt.plot(lp['labels'], lp['predictions'], color='black', linestyle='', marker='o', markersize='2', alpha=0.1)
plt.show()
```
