```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
rundir = 'out/6/'
err_df = pd.read_csv(f'{rundir}nrmse.csv')
```

```python
plt.plot(err_df['err_train'], label='nrmse_train')
plt.plot(err_df['err_val'], label='nrmse_val')
plt.yscale('log')
plt.legend()
plt.show()
```

```python
err_best = err_df['err_val'].min()
print(f'Best NRMSE: {err_best*100:.3}%')
```
